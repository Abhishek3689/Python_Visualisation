#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='region')
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,hue='event',style='region')
```

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds=sns.load_dataset('diamonds')
```

```python
diamonds
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x='price',data=diamonds,hue='cut',palette='bright')
plt.show()
```
