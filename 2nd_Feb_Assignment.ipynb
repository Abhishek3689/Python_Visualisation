#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
### Scatter Plot is a plot between two numerical variables and shows how they are related to each other
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,color='#35E11A')
plt.title("Visualisation of Scatter Plot",size=16)
plt.xlabel("X_Axis Values",fontsize=12)
plt.ylabel("Y_Axis Values",fontsize=12)
plt.grid()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
##### Use the following data:

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
### Subplot allows us to draw muliple plots in a single figure
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.plot(x1,y1,color='#198608')
plt.title("First Line Plot",size=16)
plt.grid()

plt.subplot(2,2,2)
plt.plot(x2,y2,color='#082A86')
plt.title("Second Line Plot",size=16)
plt.grid()

plt.subplot(2,2,3)
plt.plot(x3,y3,color='#7C0886')
plt.title("Third Line Plot",size=16)
plt.grid()

plt.subplot(2,2,4)
plt.plot(x4,y4,color='#86081F')
plt.title("Fourth Line Plot",size=16)
plt.grid()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
## Barplot is a plot whcih is in rectangular bar (horizontal or vertical bars) for plotting  between categorcial and numerical variables 
```

```python
## used for comparison of discrete categories
```

```python
plt.figure(figsize=(6,5))
plt.bar(company,profit,color='#18DCE9')
plt.title("Comparison of Companies Profit",size=16,color='#0B22B8')
plt.xlabel("Company Names",size=14)
plt.ylabel("Profit",size=14,color='#13630D')
plt.grid()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
### Boxplot is a plot represented in the form of BOX and gives five points summary of data (min,1st quartile,median,3rd quartile,max)
```

```python
### Boxplot usally gives if outliers is there in the dtaset
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(box1)
plt.title("Boxplot 1",size=16)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(box2,vert=False)
plt.title("Boxplot 2",size=16)
plt.grid()
plt.show()
```
